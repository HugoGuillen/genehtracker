# GeneHTracker Tutorial

You can use GeneHTracker interactively by importing the class `GHTracker` into your code. Alternatively, all functions allow the parameters `output_img` or `output_tsv` to render to file.

```python
import os
import sys
sys.path.append('.')
import genehtracker
from genehtracker import GHTracker

index_path = os.path.abspath('index')
GHT = GHTracker(index_path=index_path)
```

---
## Querying by gene name/symbol

By querying by name, the index will find all identifiers that had had the given name and then show all their historical annotation.

```python
GHT.query_genename('MALAT1')
```

The function supports approximate name search

```python
GHT.query_genename('MALATX')
```

---
## Querying by gene id

In this case, you need a Ensembl/Gencode stable id.

```python
gene_id = 'ENSG00000259484'
GHT.query_geneid(gene_id)
```

It looks that the gene `ENSG00000259484` was deprecated in Gencode version 17. To explore an alternative mapping, you can use three functions and a mutual overlap threshold (meaning, both genes have to overlap at least that fraction on both sides).

First, the history diagram:

```python
overlap_thr = 0.3
GHT.plot_history_diagram(gene_id,overlap_thr=overlap_thr)
```

If you want to check how this genes are positioned in the genome, use the function `plot_tracks`. (This is faster than loading all gencode versions in your favorite genome browser and check by hand)

```python
GHT.plot_tracks(gene_id,overlap_thr=overlap_thr)
```

Finally, you can take a look at the raw data in the graph database for the given gene.

```python
GHT.plot_history_network(gene_id,overlap_thr=overlap_thr)
```

## Batch mode

Where this functions really excel is at automatising the search for lists of genes. 

```python
gene_list = [
    'ENSG00000230544',
    'ENSG00000240453',
    'ENSG00000254100',
    'ENSG00000259484',
    'ENSG00000259758',
    'ENSG00000260940',
    'ENSG00000261496',
    'ENSG00000261685',
    'ENSG00000266584',
    'ENSG00000122952',
    'ENSG00000253827',
]
GHT.query_list(gene_list)
```

And you can call the pipeline to perform the matching search.

**Note**: in this mode the overlap threshold is not applied. Instead, all possible matches are ranked using the harmonic mean of both overlaps and the best one is returned. In case no matching was found, the script will output the names to stdout.

```python
GHT.pipeline_mapping(gene_list)
```
